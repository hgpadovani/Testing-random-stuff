# Dimensionality Reduction: seeing high dimension data

In this notebook I'll explore a few dimensionality reduction techniques. I'll try to explore them, comparing their results in terms of visualization quality in lower dimensions.

First, let's import some libs.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE
import seaborn as sns

%matplotlib inline
```

Than, let's define a function to visualize our embedding data.

```python
def plot_embedding(X, y, title=None):
    """funcao que realiza o plot de X, obtido pelo t-SNE
    Parametros
    ----------
    Inputs:
        X: vetores para plotar
        title (None): titulo opcional para o grafico
    Return:
        None
    """
    
    x_min, x_max = np.min(X, 0) - 1, np.max(X, 0)
    X = (X - x_min) / (x_max - x_min)

    plt.figure(figsize=(12,10))
    ax = plt.subplot(111)
    colors = ['red', 'blue', 'green', 'cyan', 'magenta', 'purple', 'brown', 'pink', 'gray', 'olive']
    
    for i in range(0, 10):
        ax.scatter(X[y == i, 0], X[y == i, 1], c = colors[i], label = str(i))
    ax.legend()
    if title is not None:
        ax.set_title(title)
```

Now, let's import our data. In this first task I'll use sklearn's load_dataset module, importing digits!

```python
data = load_digits()
data
```

```python
X = data.data
y = data.target

print('Shape of array: ' + str(X.shape))
print('Shape of target: ' + str(y.shape))
```

As we can see, there are 1797 samples of digits images, with 8x8 pixels, totalizing 64 dimensions. Our task is to reduce this dimension into 2 dimensions and than plot the embedding resulted by the algorithm.

But first, let's visualize a few images.

```python
# creating figure
plt.figure(figsize=(15,8))

for index, image in enumerate(np.random.randint(0, X.shape[0], 10)):
    
    ax = plt.subplot(1, 10, index+1)

    pixels = X[image,:]
    pixels = pixels.reshape((8, 8))

    ax.imshow(pixels, cmap='gray')
    
plt.show()
```

Cool! As we can see, there are digits from 0 to 9 in a 8x8 resolution image. Let's begin with the algorithms.

```python
X = StandardScaler().fit_transform(X)
```

# PCA

The first algorithm that we are going to see is Principal Component Analysis, or PCA.

```python
%time X_pca = PCA(n_components = 2, random_state = 42).fit_transform(X)
```

```python
print(X_pca.shape)
```

```python
plot_embedding(X_pca, y, "PCA of the all the digits")
```

# Isomap

Now let's see Isomap

```python
from sklearn.manifold import Isomap

%time X_isomap = Isomap(n_components = 100, n_jobs = -1).fit_transform(X)
```

```python
plot_embedding(X_isomap, y, "Isomap embedding of the digits")
```

# Locally Linear Embedding

```python
from sklearn.manifold import LocallyLinearEmbedding

%time X_lle = LocallyLinearEmbedding(n_components = 2, n_jobs = -1, random_state = 42).fit_transform(X)
```

```python
plot_embedding(X_lle, y, "Locally Linear Embedding of the digits")
```

# Spectral Embedding

```python
from sklearn.manifold import SpectralEmbedding

%time X_se = SpectralEmbedding(n_components = 2, n_jobs = -1, random_state = 42).fit_transform(X)
```

```python
plot_embedding(X_se, y, "Spectral Embedding of the digits")
```

# Multi-dimensional Scaling (MDS)

```python
from sklearn.manifold import MDS

%time X_mds = MDS(n_components = 2, n_jobs = -1, random_state = 42).fit_transform(X)
```

```python
plot_embedding(X_mds, y, "MDS Embedding of the digits")
```

#  t-distributed Stochastic Neighbor Embedding (t-SNE)

```python
from sklearn.manifold import TSNE

%time X_tsne = TSNE(n_components = 2, random_state = 42).fit_transform(X)
```

```python
plot_embedding(X_tsne, y, "TSNE Embedding of the digits")
```

# Uniform Manifold Approximation and Projection (UMAP)

```python
from umap import UMAP

%time X_umap = UMAP(n_components = 2, random_state = 42).fit_transform(X)
```

```python
plot_embedding(X_umap, y, "UAMP Embedding of the digits")
```
